```python
aps = spark.read.table('APSData')
aps = aps.withColumn("date",  ps.concat_ws("_", (ps.col("year")), ps.format_string('%02d', (ps.col("month"))))).drop('year').drop('month')
aps = aps.withColumn('dt',ps.to_timestamp(aps.date, 'yyyy_MM'))
display(aps)
```

# Category and Lens Standardization

```python
mss = spark.read.table("importancesData")
mss = mss.withColumn("date",  ps.concat_ws("_", (ps.col("year")), ps.format_string('%02d', (ps.col("month"))))).drop('year').drop('month')
mss = mss.withColumn('dt',ps.to_timestamp(mss.date, 'yyyy_MM'))
cat_df = mss.filter((ps.col('lens').isNull()) & (ps.col('sublens').isNull()) & (ps.col('category').isNotNull())).na.fill(0).drop('lens').drop('sublens')
lens_df = mss.filter((ps.col('sublens').isNull()) & (ps.col('lens').isNotNull()) & (ps.col('category').isNotNull())).na.fill(0).drop('sublens')
sublens_df = mss.filter((ps.col('sublens').isNotNull()) & (ps.col('lens').isNotNull()) & (ps.col('category').isNotNull())).na.fill(0)
```

```python
w_s_m = Window.partitionBy(sublens_df['category'],sublens_df['lens'],sublens_df['sublens'],sublens_df['date']).orderBy(sublens_df['mss'])
sublens_df = sublens_df.withColumn('percentile',ps.percent_rank().over(w_s_m)).withColumn('mss_m_percentile', ps.round(ps.col('percentile')*100,2)).drop('percentile')
display(sublens_df)
```

# Correlate Monthly MSS and CSAT Scores --> Importance

```python
#pick columns from csat df to merge with mss df to get correlation scores
merge_csat = aps.select('airline','date','CSAT','csat_m_percentile')
#pick columns from cat and lens df to merge with csat df
merge_cat_mss = cat_df.select('airline','date','category','mss','mss_m_percentile')
merge_lens_mss = lens_df.select('airline','date','lens','mss','mss_m_percentile')
merge_sublens_mss = sublens_df.select('airline','date','sublens','mss','mss_m_percentile')
#make cat and lens correlation dfs
cat_corr = merge_csat.join(merge_cat_mss, on = ['airline','date'])
lens_corr = merge_csat.join(merge_lens_mss, on = ['airline','date'])
sublens_corr = merge_csat.join(merge_sublens_mss, on = ['airline','date'])

#define correlation udf
findCorrUDF = udf(lambda x, y: float(stats.pearsonr(x, y)[0]), DoubleType())
```

```python
cat_csat_series = cat_corr.withColumn("csat_series", ps.collect_list("csat").over(Window.partitionBy("airline","category").orderBy("date")))
cat_csat_series = cat_csat_series.groupby("airline","category").agg(ps.max("csat_series").alias("csat_series"))

cat_mss_series = cat_corr.withColumn("mss_series", ps.collect_list("mss").over(Window.partitionBy("airline","category").orderBy("date")))
cat_mss_series = cat_mss_series.groupby("airline","category").agg(ps.max("mss_series").alias("mss_series")).withColumnRenamed("airline", "airline_b")\
                  .withColumnRenamed("category", "category_b")

full_cat_series = cat_csat_series.join(cat_mss_series, on = [cat_csat_series['airline'] == cat_mss_series['airline_b'],
                                                               cat_csat_series['category'] == cat_mss_series['category_b']], how = 'inner').drop('airline_b').drop("category_b")
full_cat_series = full_cat_series.withColumn("CatMssCsatCorr_raw", findCorrUDF("csat_series", "mss_series")).withColumn('CatMssCsatCorr(Importance)', ps.round(ps.col('CatMssCsatCorr_raw'),2))\
                  .drop('CatMssCsatCorr_raw').drop('csat_series').drop('mss_series')
display(full_cat_series)
```

```python
lens_csat_series = lens_corr.withColumn("csat_series", ps.collect_list("CSAT").over(Window.partitionBy("airline","lens").orderBy("date")))
lens_csat_series = lens_csat_series.groupby("airline","lens").agg(ps.max("csat_series").alias("csat_series"))

lens_mss_series = lens_corr.withColumn("mss_series", ps.collect_list("mss").over(Window.partitionBy("airline","lens").orderBy("date")))
lens_mss_series = lens_mss_series.groupby("airline","lens").agg(ps.max("mss_series").alias("mss_series")).withColumnRenamed("airline", "airline_b")\
                  .withColumnRenamed("lens", "lens_b")

full_lens_series = lens_csat_series.join(lens_mss_series, on = [lens_csat_series['airline'] == lens_mss_series['airline_b'],
                                                               lens_csat_series['lens'] == lens_mss_series['lens_b']], how = 'inner').drop('airline_b').drop("lens_b")
full_lens_series = full_lens_series.withColumn("LensMssCsatCorr_raw", findCorrUDF("csat_series", "mss_series")).withColumn('LensMssCsatCorr(Importance)', ps.round(ps.col('LensMssCsatCorr_raw'),2))\
                  .drop('LensMssCsatCorr_raw').drop('csat_series').drop('mss_series')
display(full_lens_series)
```

```python
sublens_csat_series = sublens_corr.withColumn("csat_series", ps.collect_list("csat").over(Window.partitionBy("airline","sublens").orderBy("date")))
sublens_csat_series = sublens_csat_series.groupby("airline","sublens").agg(ps.max("csat_series").alias("csat_series"))

sublens_mss_series = sublens_corr.withColumn("mss_series", ps.collect_list("mss").over(Window.partitionBy("airline","sublens").orderBy("date")))
sublens_mss_series = sublens_mss_series.groupby("airline","sublens").agg(ps.max("mss_series").alias("mss_series")).withColumnRenamed("airline", "airline_b")\
                  .withColumnRenamed("sublens", "sublens_b")

full_sublens_series = sublens_csat_series.join(sublens_mss_series, on = [sublens_csat_series['airline'] == sublens_mss_series['airline_b'],
                                                               sublens_csat_series['sublens'] == sublens_mss_series['sublens_b']], how = 'inner').drop('airline_b').drop("sublens_b")
full_sublens_series = full_sublens_series.withColumn("SublensMssCsatCorr_raw", findCorrUDF("csat_series", "mss_series"))\
                  .withColumn('SublensMssCsatCorr(Importance)', ps.round(ps.col('SublensMssCsatCorr_raw'),2))\
                  .drop('SublensMssCsatCorr_raw').drop('csat_series').drop('mss_series')
display(full_sublens_series)
```

```python
#helper funcs to write to excel
def check_copyfile(copy_path: str, writing_path: str, databricks_write: bool):
    if databricks_write:
        copyfile(copy_path, writing_path)

def writing_check(writing_path: str, databricks_write: bool, writing_name: str) -> str:
    if databricks_write:
        copy_path = "/tmp/" + writing_name + ".xlsx"
    else:
        copy_path = writing_path
    return copy_path

def df_to_file(result_dict: dict, writing_path: str, databricks_write: bool = True) -> str:
  copied_path = writing_check(writing_path, databricks_write, "deduplication_result")
  with pd.ExcelWriter(copied_path) as writer:
      for key in result_dict.keys():
        result_dict[key].to_excel(writer, sheet_name=key)
  check_copyfile(copied_path, writing_path, databricks_write)
  return writing_path

pandas_dict = {'csat': aps.toPandas(), 'cat_mss' : cat_df.toPandas(), 'lens_mss': lens_df.toPandas(), 'sublens_mss': sublens_df.toPandas(), 
               'cat_corr': full_cat_series.toPandas(), 'lens_corr': full_lens_series.toPandas(), 'sublens_corr': full_sublens_series.toPandas()}

output_path = '/bsinsights/Raim'
output_filename = 'covid_CSAT_MSS_IMP'
out_file_prefix =  output_path + "/" + output_filename
writing_path = "/dbfs/mnt/" + out_file_prefix + ".xlsx"
df_to_file(pandas_dict,writing_path, True)
```
